```python
#Importing packages.
from bs4 import BeautifulSoup, SoupStrainer
import requests
from collections import Counter
import time
```

```python
#Setting up Beautiful Soup for the Taylor Swift Artist page.
url = "https://www.azlyrics.com/t/taylorswift.html"
page = requests.get(url)    
data = page.text
soup = BeautifulSoup(data)
links = []
for link in soup.find_all('a'):
    links.append(link)
```

```python
#Finding the links of all of the links on the above page which are Taylor Swift songs.
cleanlinks=[]
cleanerlinks=[]
fulllinks=[]
for link in links:
    if '/lyrics/taylorswift' in link.get('href', ''):
        cleanlinks.append(link.get('href', ''))
for link in cleanlinks:
    cleanerlinks.append(link[3:])
for link in cleanerlinks:
    fulllinks.append('https://www.azlyrics.com/'+link)
```

```python
#Getting all text from each song.
all=''
for url in fulllinks:
    page = requests.get(url)    
    data = page.text
    soup = BeautifulSoup(data)
    alltext=[]
    for text in soup.find_all('div'):
        alltext.append(text)
    temp = alltext[21].text
    all = all + temp
    print("Done, sleeping")
    time.sleep(30)
```

```python
#Cleaning.
all = all.replace("\r", "")
all = all.replace("\n\n","\n")
all = all.replace("\n"," ")  
all = all.replace("[Album version line:]"," ")
all = all.replace("[Music video version line:]"," ")
all = all.replace(",","")
all = all.replace("...","")
all = all.lower()
```

```python
#Getting a string containing every line of every song from the links we found earlier.
raw_lines=''
for url in fulllinks:
    page = requests.get(url)    
    data = page.text
    soup = BeautifulSoup(data)
    alltext=[]
    for text in soup.find_all('div'):
        alltext.append(text)
    temp = alltext[21].text
    raw_lines = raw_lines + temp
    print("Done, sleeping")
    time.sleep(30)
```

```python
#Cleaning the string found above.
raw_lines = raw_lines.replace("\r", "")
raw_lines = raw_lines.replace("\n\n","\n")
raw_lines = raw_lines.replace("[Album version line:]"," ")
raw_lines = raw_lines.replace("[Music video version line:]"," ")
raw_lines = raw_lines.replace(",","")
raw_lines = raw_lines.replace("...","")
raw_lines = raw_lines.lower()
```

```python
#Splitting at newline characters
lines = raw_lines.split("\n")
lines = list(filter(None, lines))
lines
```

```python
#Writing to file, using utf-8 encoding because of the presence of non ASCII characters in the text 
with open('all_lyrics.txt', 'w', encoding='utf-8') as f:
    for item in lines:
        f.write("%s\n" % item)
        
#import io
#with io.open('all_lyrics.txt', "w", encoding="utf-8") as f:
#    f.write(html)
```
